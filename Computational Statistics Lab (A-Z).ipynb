```python
import pandas as pd
from pandas import DataFrame
from sklearn import linear_model
import statsmodels.api as sm
```

```python
import pandas as pd
from pandas import DataFrame
from sklearn import linear_model
import statsmodels.api as sm
```

```python
X=df[['Interest_Rate','Unemployment_Rate']]
Y=df['Stock_Index_Price']
```

```python
#K-Means Clustering: K-Means Clustering is a type of clustering method that sudivides a single cluster or a collection of data 
#points into k different clusters or groups.It is an unsupervized learning algorithm which groups the unlabeled dataset into 
#different clusters.It is a centroid based algorithm where each cluster is associated with a centroid.The main aim of this 
#algorithm is to minimize the sum of distances between the data points and their corresponding clusters.
#Here k defines the no. of predefined clusters that needs to be created in the process, for ex- if k = 2 there will be two 
#clusters, if k=3 there will be 3 clusters and so on.
#It is an iteretive algorithm that divides the unlabeled data into k different clusters in such a way that each dataset belongs 
#to only one grp that has similar properties.
```

```python
from pandas import DataFrame
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
```

```python
Data ={'x':[185,170,168,179,182,188,180,180,183,180,180,177],
      'y':[72,56,60,68,72,77,71,70,84,88,67,76]}
df=DataFrame(Data,columns=['x','y'])
print("The shape of data is:",df.shape)
df
```

```python
#Elbow method:
#Choosing the right no. of clusters is an important part of this algorithm.
#An ideal way to figure out the right no. of cluster is to calculate the Within Cluster Sum of Squares(WCSS)
#A commonly used method for finding the optimal value of k(no. of clusters) is Elbow Method.
#In this method, we are actually verifying the no. of clusters(k) generally from 1 to 10.
#For each value of k we are calculating WCSS.
#WCSS is the sum of squared distance between each point and the centroid in a cluster.
#When we plot the WCSSS value with kValue the plot looks like an elbow.
#As the no. of clusters increases the wcss value will start to decrease.
```

```python
wcss=[]      #Initializes an elbow list to store the value
for i in range (1,13):
    km=KMeans(n_clusters=i)
    km.fit_predict(df)
    wcss.append(km.inertia_)
wcss
```

```python
plt.plot(range(1,13),wcss)
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")
plt.title("The Elbow Point graph for optimal values of k in K-Means Clustering")
plt.show()
```

```python
#K-Means Clustering
my_centroids=np.array([[185,72],[170,56],[168,60]])
kmeans=KMeans(n_clusters=3,init=my_centroids).fit(df)
centroids=kmeans.cluster_centers_
print(centroids)
```

```python
X=df.iloc[:,:].values
y_kmeans=kmeans.fit_predict(X)
y_kmeans
```

```python
plt.scatter(X[y_kmeans==0,0],X[y_kmeans==0,1],s=60,c='lime',label='Cluster 1')      #Scatter plot for data points in cluster 1
plt.scatter(X[y_kmeans==1,0],X[y_kmeans==1,1],s=60,c='deeppink',label='Cluster 2')  #Scatter plot for data points in cluster 2
plt.scatter(X[y_kmeans==2,0],X[y_kmeans==2,1],s=60,c='chocolate',label='Cluster 3') #Scatter plot for data points in cluster 3
plt.scatter(centroids[:,0], centroids[:,1],c='purple',s=100,marker='*',label='Centroids') #Scatter plot for centroid
plt.legend(loc='lower right')        #Adding legend
plt.show()
```

```python
#PCA is a dimensionality reduction method that is often use to reduce the diamensionality of large data set by transforming a 
#large set of variables into a smaller one,that still contains most of the information in the large set.That is dimentionality 
#of the data reduced while retaining most of it's original variability.
#In this technique, variables are transformed into a new set of variables which are linear combination of original variables.
#This new set of variables are known as principal components.The 1st principal component captures the most variation in the data
#but the 2nd principal component captures the maximum variance that is orthogonal to the 1st principal components and so on.
#In PCA, no. of features = no. of components.
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("iris.csv")
df
```

```python
#Dividing data into labels and feature sets
X = df.iloc[:, 0:4].values
y = df.iloc[:,4].values
```

```python
#Split the data into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
#Feature scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#Performing Linear Discriminant Analysis(LDA is applied to reduce the dimentionality of the data to 1 component)
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
#Initialize the LDA model
lda = LinearDiscriminantAnalysis(n_components=1)
X_train_lda = lda.fit_transform(X_train, y_train)
X_test_lda = lda.transform(X_test)
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(max_depth=2,random_state=0)
classifier.fit(X_train_lda,y_train)                            #X_train_lda: Training data after applying LDA.
y_pred= classifier.predict(X_test_lda)                         #X_test_lda: Test data after applying the same LDA transformation
#as used for the training data.This ensures that the test data is in the same reduced feature space as the training data.
```

```python
#Evaluate the performance of the classifier
conf_matrix = confusion_matrix(y_test,y_pred)
accuracy = accuracy_score(y_test,y_pred)
```

```python
print("Confusion Matrix:\n", conf_matrix)
print("Accuracy:",accuracy)
```

```python
#The confusion matrix shows the no. of true positive, true negative, false positive and false negative predictions; while 
#acccuracy represents the proportion of correctly predicted instances.
```
